#Distribuição Amostral e Intervalo de Confiança
Nossa amostra de 4271 valores representa aproximadamente 20% do total de cientista de dados brasileiros, que ao pesquisar no linkedin dá um total de aproximadamente 18 mil. Então temos uma amostra de tamanho suficiente para análise.

```python
#Calculando média de salarios :
media_amostral = salarios.mean()
media_amostral
```

```python
#Calculando desvio padrão:
desvio_amostral = salarios.std()
desvio_amostral
```

```python
from scipy import stats
```

```python
#Calculo erro padrão:
erro_padrao = stats.sem(salarios)
erro_padrao
```

```python
#Calculando intervalo de confiança da média salarial:
intervalo_confianca = stats.t.interval(nivel_confianca, tamanho_amostra - 1, loc=media_amostral, scale=erro_padrao)
intervalo_confianca
#Ou seja, podemos afirmar que a média salarial de cientista de dados do brasil gira em torno de 9.131 a 9.562 reais para um nível de confiança de 95%.
```

#Featuring engineering
Feature engineering, ou engenharia de recursos em português, é o processo de transformar dados brutos em recursos (features) que sejam mais úteis para modelos de aprendizado de máquina (machine learning). Isso envolve criar, selecionar e transformar variáveis existentes, ou criar novas a partir de combinações de variáveis.

##1)Combinação de Variáveis

```python
#Iremos criar novos valores a partir de outros existentes no dataframe.
#Remodelar os valores de nível a partir dos dados da coluna pessoa gestora.

#Primeiro, relembrar nome das colunas:
dados_sem_null.columns
```

```python
#Criando função para tratar valores :
def preencher_nivel(gestor, nivel):
  if gestor == 1.0:
    return "Pessoa Gestora"
  else:
    return nivel
```

```python
#Inserindo resultados da função numa nova coluna NOVO_NIVEL:
dados_sem_null["NOVO_NIVEL"] = dados_sem_null.apply(lambda x: preencher_nivel(x["GESTOR?"], x["NIVEL"]), axis=1)
```

```python
#Calculando variáveis discretas e correlação entre as tabelas cruzadas:

#Importando biblioteca
from scipy.stats import chi2_contingency

#Declarando função para coefiente de cramer:
#O coeficiente de Cramer (também conhecido como V de Cramer) é uma medida de associação entre duas variáveis categóricas.
#Ele é usado para quantificar a força da relação entre duas variáveis nominais ou ordinais em uma tabela de contingência.

def cramer_coeficiente(coluna1, coluna2):
  #Criando tabela cruzada: (cruza valores da coluna 1 e coluna2 para cálculo da correlação)
  tabela_cruzada = np.array(pd.crosstab(coluna1, coluna2))
  #calculo do teste qui-quadrado usando função chi2_contingency:
  chi2 = chi2_contingency(tabela_cruzada)[0] #teste que avalia associação e independência entre variáveis
  #criando variáveis soma e mini:
  soma = np.sum(tabela_cruzada) #armazena a soma e todos elementos da tabela cruzada
  mini = min(tabela_cruzada.shape) - 1 #graus de liberdade do qui-quadrado. Retorna o menor valor entre o número de linhas e o número de colunas da tabela cruzada.
  #calculando coeficiente de cramer
  cramer = np.sqrt(chi2/(soma*mini)) #np.sqrt= função pra calcular raiz quadrada; fórmula = raiz de (qui quadradi/ (soma * mini))
  return tabela_cruzada, chi2, soma, mini, cramer
```

```python
#Usando a função cramer_coeficiente para calcular o coeficiente de cramer entre duas variáveis categóricas:
cramer_coeficiente(dados_atual["COR/RACA/ETNIA"], dados_atual["NIVEL DE ENSINO"])

#Resultados:
#tabela cruzada entre coluna 1 ["COR/RACA/ETNIA"]e coluna 2 ["NIVEL DE ENSINO"]
#teste qui-quadrado: 51.89633710573608
#soma: somando todos os números da tabela cruzada obtemos um total de 4273
#mini: a tabela é 7 colunas x 7 linhas e por isso mini é 7-1=6
#coeficiente de cramer:Deu 0.4 o que indica associação moderada
#< 0.1: Associação muito fraca
#0.1 a 0.3: Associação fraca
#0.3 a 0.5: Associação moderada
#>= 0.5: Associação forte
```
